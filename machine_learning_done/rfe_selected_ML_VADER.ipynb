```python
from sklearn.naive_bayes import GaussianNB
import os
import json
import pandas as pd
import numpy as np
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
# Set the working directory
os.chdir('C:/Users/jesse/OneDrive/Documenten/Thesis/amazon_code/dataframes_done')

# Read the DataFrame from a JSON file
df = pd.read_json('VADER_df.json', orient='records')

# Assign the independent variables to X and dependent variable to y
X = df[['...']]
y = df["Label"] # label (0 or 1) is the dependent variable

# Split the data into training and test sets
train_size = 0.8
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=42)

print('Entire dataset size:', len(df))
print('Train dataset size:', len(X_train))
print('Test dataset size:', len(X_test))

# Implementing SVM
svm1 = SVC(kernel='linear')
svm1.fit(X_train, y_train)
predictions = svm1.predict(X_test)

# Evaluate
svm_accuracy = accuracy_score(y_test, predictions)*100
svm_precision = precision_score(y_test, predictions)* 100
svm_recall = recall_score(y_test, predictions)*100
svm_f1_score = f1_score(y_test, predictions)*100

# Confusion matrix
cm = confusion_matrix(y_test, predictions, labels=svm1.classes_)
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=svm1.classes_) 
display.plot()
print(svm_accuracy)
print(svm_precision)
print(svm_recall)
print(svm_f1_score)

###################################################################################
# Implementing GNB
gnb1 = GaussianNB()
gnb1.fit(X_train, y_train)
prediction = gnb1.predict(X_test)

# Evaluate
gnb_accuracy = accuracy_score(y_test, prediction)*100
gnb_precision = precision_score(y_test, prediction)* 100
gnb_recall = recall_score(y_test, prediction)*100
gnb_f1_score = f1_score(y_test, prediction)*100

# Confusion matrix
cm = confusion_matrix(y_test, prediction, labels=gnb1.classes_)
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=gnb1.classes_) 
display.plot()
print('GNB Evaluation:')
print(gnb_accuracy)
print(gnb_precision)
print(gnb_recall)
print(gnb_f1_score)
```
